```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_cars = pd.read_csv('bar_chart_data.csv')
df_cars
```

```python
# the default size is 6.4 x 4.8
plt.figure(figsize = (9, 6))
plt.bar(x = df_cars['Brand'], 
        height = df_cars['Cars Listings'], 
        color = 'midnightblue')
plt.xticks(rotation = 45, fontsize =13)
plt.yticks(fontsize = 13)
plt.title('Cars Listings by Brand', 
          fontsize = 16, 
          fontweight ='bold')
plt.ylabel('Number of Listings', 
           fontsize = 13)
plt.show()
plt.savefig("Cars Bar.png")
```

## Pie Chart

```python
df_fuel_type = pd.read_csv('pie_chart_data.csv')
df_fuel_type
```

```python
sns.set_palette('colorblind')
```

```python
# autopct take as string or function to label the wedges
plt.figure(figsize = (10, 8))
plt.pie(df_fuel_type['Number of Cars'], 
        labels = df_fuel_type['Engine Fuel Type'].values, 
        autopct = '%.2f%%', 
        textprops = {'size': 'x-large', 
                     'fontweight': 'bold', 
                     'rotation': 30, 
                     'color': 'w'})
plt.legend()
plt.title('Cars by Engine Fuel Type', 
          fontsize = 18, 
          fontweight = 'bold')
plt.show()
```

## Stacked Area Chart

```python
sns.set()
```

```python
df_fuel_engine_types = pd.read_csv('stacked_area_chart_data.csv')
df_fuel_engine_types
```

```python
colors = ['#011638', '#7e2987', '#ef2026']
labels = ['Diesel', 'Petrol', 'Gas']
sns.set_style('white') # background
plt.figure(figsize = (12, 6))
plt.stackplot(df_fuel_engine_types['Year'], 
              df_fuel_engine_types['Diesel'], 
              df_fuel_engine_types['Petrol'], 
              df_fuel_engine_types['Gas'], 
              colors = colors, edgecolor = 'none')
plt.xticks(df_fuel_engine_types['Year'], rotation = 45)
plt.legend(labels = labels, loc = 'upper left')
plt.ylabel('Number of Cars', fontsize =13)
plt.title("Popularity of Engine Fuel Types (1982 - 2016)", fontsize = 14, weight = 'bold')
sns.despine() # remove borders
plt.show()
```

## Scatter Plot

```python
df_real_estate = pd.read_csv('scatter_data.csv')
df_real_estate
```

```python
plt.figure(figsize = (12, 8))
scatter = plt.scatter(df_real_estate['Area (ft.)'], 
            df_real_estate['Price'], 
            alpha = 0.6, 
            c = df_real_estate['Building Type'], 
            cmap = 'viridis')

# points to the memory where the first element of scatter's legend_element is stored
plt.legend(*scatter.legend_elements(), 
           loc = 'upper left', 
           title = 'Building Type')
plt.title("Relationship between Area and Price of California Real Estate", 
          fontsize = 14,
          fontweight = 'bold')
plt.xlabel("Area (sq .ft.)", weight = 'bold')
plt.ylabel("Price (000's of $)")
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'Price', 
                y = 'Area (ft.)', 
                data = df_real_estate,  # Pass the data as a DataFrame
                hue ='Building Type', 
                palette = ['black', 'darkblue', 'purple', 'pink', 'white'], 
                s = 100)
plt.title("Relationship between Area and Price of California Real Estate", 
          fontsize = 14, 
          fontweight = 'bold')
plt.xlabel("Area (sq .ft.)", weight = 'bold')
plt.ylabel("Price (000's of $)")
plt.grid(True)
plt.show()
```

## Regression Plot

```python
df_ad_expenditure = pd.read_csv('scatter_plot_ii.csv')
df_ad_expenditure
```

### sns.regplot() accepts variety of formats for the X and Y variables, such as numpy arrays, pandas series, and others. lmplot, can only support X and Y in the form of strings but wider range of features.

```python
sns.set(rc = {'figure.figsize': (9,6)})
sns.regplot(x = 'Budget', 
            y = 'Sales', 
            data = df_ad_expenditure, 
            scatter_kws = {'color': 'k'}, 
            line_kws = {'color': 'red'})
plt.xlabel("Ad Expenditure in (000's Units)")
plt.ylabel("Sales in (000's)")
plt.show()
```

```python
sns.lmplot(x = "Budget", 
           y = "Sales", 
           data = df_ad_expenditure, 
           height = 10, 
           scatter_kws = {'color': 'k'}, 
           line_kws = {'color': 'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", 
          fontsize = 14, 
          weight = "bold")
plt.show()
```

## Bar and Line Chart

```python
from matplotlib.ticker import PercentFormatter
```

```python
df_kdnuggets = pd.read_csv('bar_line_chart_data.csv')
df_kdnuggets
```

```python
sns.reset_orig()
```

```python
# Create a figure and axes for the plot
fig, ax = plt.subplots(figsize=(10, 7))

# Create a bar plot on the left y-axis (ax)
ax.bar(df_kdnuggets['Year'], 
       df_kdnuggets['Participants'], 
       color='#3376BD')

# Set the label for the left y-axis
ax.set_ylabel("Number of Participants", 
              weight='bold')

# Customize the tick parameters of the left y-axis
ax.tick_params(axis='y', 
               width=2, 
               labelsize='large')
#-------------------------------------------------------
# Create a twin y-axis on the right side (ax1)
ax1 = ax.twinx()

# Set the y-axis limits for the right y-axis (ax1)
ax1.set_ylim(0, 1)

# Format the ticks on the right y-axis as percentages
ax1.yaxis.set_major_formatter(PercentFormatter(xmax=1.0))

# Create a line plot on the right y-axis (ax1)
ax1.plot(df_kdnuggets['Year'], 
          df_kdnuggets['Python Users'], 
          color='black', 
          marker='D')

# Set the label for the right y-axis
ax1.set_ylabel('Python Users', 
               color='black', 
               weight='bold')

# Customize the tick parameters of the right y-axis
ax1.tick_params(axis='y', 
                color='black', 
                width=2, 
                labelsize='large')

# Show the plot
plt.show()
```
